Use SymPy to solve again the problem of throwing a ball straight up with an initial speed $v_0 = 5 m/s$. Evaluate symbolically the maximum height $y_max$ reached by the ball (neglecting air resistance), and how long it takes to reach this maximum height.



## Solution

Using 

$$y = v_0t - \frac{1}{2}gt^2$$

e 

$$v(t) = v_0 - gt$$

```python
from sympy import symbols, solve
```

```python
v0, t, g = symbols('v0, t, g', real = True)
v0, t, g
```

Solve equation for tmax symbolically using solve() function

```python
tmax = solve(v0 - g*t, t)[0]
print('Time to reach max height:')
tmax
```

Max height reached

```python
y = v0*t - g*t**2/t
y
```

substitute tmax in y(t) to find ymax

```python
print('Max height reached:')
y.subs(t,tmax)
```
