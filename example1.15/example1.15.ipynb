# Example of plot and numerical evaluation

Write a code to evaluate and plot the kinematic equation (1.6.1), from $t = 0$ to $t = 1$ s with $vo = 5\, m/s$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v0, g = 5, 9.8
```

```python
t = np.linspace(0, 1.03, 100)
y = np.array([v0*u - g*u**2/2 for u in t])
```

```python
plt.plot(t, y, 'r--')
plt.title('Ball thrown straight up')
plt.ylabel('Vertical distance y(t)')
plt.xlabel('Time t [s]')
ymax = np.max(y)
tmax = t[np.argmax(y)]

print('Max height reached = ',f'{ymax:.4f}', ' m')
print('Time to reach ymax = ',f'{tmax:.4f}',' s')
plt.show()
```
